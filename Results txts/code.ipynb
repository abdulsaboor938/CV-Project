### code for reading all data afrom files and saving into csv's

```python
import pandas as pd
import re
from datetime import datetime

# Initialize lists to store data
epochs = []
f1_scores = []
kendal = []
spear = []
total_times = []
model = []

# origsumlogs.txt
with open ('origsumlogs.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if 'max_epoch' in line:
            epochs.append(int(line.split('max_epoch=')[1].split(',')[0].strip()))
        if 'F1-score' in line:
            f1_scores.append(float(line.split('F1-score ')[1].split(',')[0].split('%')[0].strip()))
        if 'Kendal' in line:
            kendal.append(float(line.split('Kendal ')[1].split(',')[0].strip()))
        if 'Spear' in line:
            spear.append(float(line.split('Spear ')[1].split(',')[0].strip()))
        if 'Total elapsed time' in line:
            time = line.split('Total elapsed time (h:m:s):')[1].strip()
            # convert to seconds
            time = datetime.strptime(time, '%H:%M:%S')
            total_times.append(time.hour * 3600 + time.minute * 60 + time.second)
```
