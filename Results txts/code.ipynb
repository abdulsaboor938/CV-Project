### code for reading all data afrom files and saving into csv's

```python
import pandas as pd
import re
from datetime import datetime

# Initialize lists to store data
epochs = []
f1_scores = []
kendal = []
spear = []
total_times = []
model = []

# origsumlogs.txt
with open ('origsumlogs.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if 'max_epoch' in line:
            epochs.append(int(line.split('max_epoch=')[1].split(',')[0].strip()))
        if 'F1-score' in line:
            f1_scores.append(float(line.split('F1-score ')[1].split(',')[0].split('%')[0].strip()))
        if 'Kendal' in line:
            kendal.append(float(line.split('Kendal ')[1].split(',')[0].strip()))
        if 'Spear' in line:
            spear.append(float(line.split('Spear ')[1].split(',')[0].strip()))
        if 'Total elapsed time' in line:
            time = line.split('Total elapsed time (h:m:s):')[1].strip()
            # convert to seconds
            time = datetime.strptime(time, '%H:%M:%S')
            total_times.append(time.hour * 3600 + time.minute * 60 + time.second)
```

```python
# origtvlogs.txt
origtv_f1_scores = []
origtv_kendal = []
origtv_spear = []
origtv_total_times = []

with open ('origtvlogs.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if 'F1-score' in line:
            origtv_f1_scores.append(float(line.split('F1-score ')[1].split(',')[0].split('%')[0].strip()))
        if 'Kendal' in line:
            origtv_kendal.append(float(line.split('Kendal ')[1].split(',')[0].strip()))
        if 'Spear' in line:
            origtv_spear.append(float(line.split('Spear ')[1].split(',')[0].strip()))
        if 'Total elapsed time' in line:
            time = line.split('Total elapsed time (h:m:s):')[1].strip()
            # convert to seconds
            time = datetime.strptime(time, '%H:%M:%S')
            origtv_total_times.append(time.hour * 3600 + time.minute * 60 + time.second)

# trimsunlogs.txt
trimsum_f1_scores = []
trimsum_kendal = []
trimsum_spear = []
trimsum_total_times = []

with open ('trimsumlogs.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if 'F1-score' in line:
            trimsum_f1_scores.append(float(line.split('F1-score ')[1].split(',')[0].split('%')[0].strip()))
        if 'Kendal' in line:
            trimsum_kendal.append(float(line.split('Kendal ')[1].split(',')[0].strip()))
        if 'Spear' in line:
            trimsum_spear.append(float(line.split('Spear ')[1].split(',')[0].strip()))
        if 'Total elapsed time' in line:
            time = line.split('Total elapsed time (h:m:s):')[1].strip()
            # convert to seconds
            time = datetime.strptime(time, '%H:%M:%S')
            trimsum_total_times.append(time.hour * 3600 + time.minute * 60 + time.second)

# trimtvlogs.txt
trimtv_f1_scores = []
trimtv_kendal = []
trimtv_spear = []
trimtv_total_times = []

with open ('trimtvlogs.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if 'F1-score' in line:
            trimtv_f1_scores.append(float(line.split('F1-score ')[1].split(',')[0].split('%')[0].strip()))
        if 'Kendal' in line:
            trimtv_kendal.append(float(line.split('Kendal ')[1].split(',')[0].strip()))
        if 'Spear' in line:
            trimtv_spear.append(float(line.split('Spear ')[1].split(',')[0].strip()))
        if 'Total elapsed time' in line:
            time = line.split('Total elapsed time (h:m:s):')[1].strip()
            # convert to seconds
            time = datetime.strptime(time, '%H:%M:%S')
            trimtv_total_times.append(time.hour * 3600 + time.minute * 60 + time.second)

# cnnsumlogs.txt
cnnsum_f1_scores = []
cnnsum_kendal = []
cnnsum_spear = []
cnnsum_total_times = []

with open ('cnnsumlogs.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if 'F1-score' in line:
            cnnsum_f1_scores.append(float(line.split('F1-score ')[1].split(',')[0].split('%')[0].strip()))
        if 'Kendal' in line:
            cnnsum_kendal.append(float(line.split('Kendal ')[1].split(',')[0].strip()))
        if 'Spear' in line:
            cnnsum_spear.append(float(line.split('Spear ')[1].split(',')[0].strip()))
        if 'Total elapsed time' in line:
            time = line.split('Total elapsed time (h:m:s):')[1].strip()
            # convert to seconds
            time = datetime.strptime(time, '%H:%M:%S')
            cnnsum_total_times.append(time.hour * 3600 + time.minute * 60 + time.second)

# cnntvlogs.txt
cnntv_f1_scores = []
cnntv_kendal = []
cnntv_spear = []
cnntv_total_times = []

with open ('cnntvlogs.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        if 'F1-score' in line:
            cnntv_f1_scores.append(float(line.split('F1-score ')[1].split(',')[0].split('%')[0].strip()))
        if 'Kendal' in line:
            cnntv_kendal.append(float(line.split('Kendal ')[1].split(',')[0].strip()))
        if 'Spear' in line:
            cnntv_spear.append(float(line.split('Spear ')[1].split(',')[0].strip()))
        if 'Total elapsed time' in line:
            time = line.split('Total elapsed time (h:m:s):')[1].strip()
            # convert to seconds
            time = datetime.strptime(time, '%H:%M:%S')
            cnntv_total_times.append(time.hour * 3600 + time.minute * 60 + time.second)
```

```python
# creating a dataframe
data = {
    'Epochs': epochs,
    'origsum F1-score': f1_scores,
    'trimsum_f1_scores': trimsum_f1_scores,
    'cnnsum_f1_scores': cnnsum_f1_scores,
    'origtv_f1_scores': origtv_f1_scores,
    'trimtv_f1_scores': trimtv_f1_scores,
    'cnntv_f1_scores': cnntv_f1_scores,
    'origsum Kendal': kendal,
    'trimsum_kendal': trimsum_kendal,
    'cnnsum_kendal': cnnsum_kendal,
    'origtv_kendal': origtv_kendal,
    'trimtv_kendal': trimtv_kendal,
    'cnntv_kendal': cnntv_kendal,
    'origsum Spear': spear,
    'trimsum_spear': trimsum_spear,
    'cnnsum_spear': cnnsum_spear,
    'origtv_spear': origtv_spear,
    'trimtv_spear': trimtv_spear,
    'cnntv_spear': cnntv_spear,
    'origsum_total_times': total_times,
    'trimsum_total_times': trimsum_total_times,
    'cnnsum_total_times': cnnsum_total_times,
    'origtv_total_times': origtv_total_times,
    'trimtv_total_times': trimtv_total_times,
    'cnntv_total_times': cnntv_total_times,
    }

df = pd.DataFrame(data)
df.to_csv('results.csv', index=False)
```
